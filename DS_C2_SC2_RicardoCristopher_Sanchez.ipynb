**Se importa el framework de pandas**


```python
import pandas as pd
```

**A) Se extrae la información del archivo de datos de vacunación anti COVID**

```python
vacunas = pd.read_csv('country_vaccinations.csv')
vacunas.head(5)
```

 **B)Se muestra la estructura y los tipos de datos del framework**


```python
vacunas.info()
```

```python
# se puede ver que la columna date es de tipo objet, por lo que se cambiará a fecha (datetime64)
vacunas['date'] = pd.to_datetime(vacunas['date'])
vacunas.info()
```

**C) Suma de Vacunas por compañia**

```python
vacunas['vaccines'].head(5)
```

```python
# se hace un split por coma de la columna vaccines y se agrega al datafrmes; luego se cuenta a partir del explode
vacunas['vaccines_split'] = vacunas['vaccines'].str.split(', ')
vacunas_exploded = vacunas.explode('vaccines_split')
conteo = vacunas_exploded['vaccines_split'].value_counts()
print(conteo)
```

**D) Cantidad de vacunas aplicadas en todo el mundo**

```python
# Sumar todas las vacunas aplicadas
total_vacunas = vacunas['daily_vaccinations'].sum()
print ("El total mundial de vacunas es de:", total_vacunas)
```

**E) Calcular el promedio de vacunas aplicadas por país**

```python
pais = vacunas.groupby('country')['daily_vaccinations'].mean()
print(pais)
```

**F) Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo**

```python
fecha = vacunas[vacunas['date'] == '2021-01-29']
# sumatoria
ene = fecha['daily_vaccinations'].sum()

print("Total de vacunas aplicadas el 2021-01-29: ", ene)
```

**G) Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.**

```python
conDiferencias = vacunas
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
conDiferencias['diferencias'].tail(5)
```

**H) Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua**

```python
fecha_min = vacunas['date'].min()
fecha_max = vacunas['date'].max()

# Calcular la diferencia
periodo_total = fecha_max - fecha_min
print(periodo_total)
```

**I) Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).**

```python
# Crear el nuevo DataFrame
conCantidad = vacunas

conCantidad['canVac'] = conCantidad['vaccines'].str.split(',').str.len()
conCantidad.head(5)
```

**J. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.**

```python
antes20 = vacunas[vacunas['date'] < '2020-12-20']
print(antes20.head(5))
print(antes20['date'].head(5))
```

**K) Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.**

```python
# Crear el DataFrame pfizer con registros que contengan Pfizer
pfizer = vacunas[vacunas['vaccines'].str.contains('Pfizer')]
pfizer.head(5)
```

**L) Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter.**

```python
excel = pd.ExcelWriter('resultadosReto.xlsx')

conDiferencias.to_excel(excel, sheet_name='conDiferencias', index=False)
conCantidad.to_excel(excel, sheet_name='conCantidad', index=False)
antes20.to_excel(excel, sheet_name='antes20', index=False)
pfizer.to_excel(excel, sheet_name='pfizer', index=False)

excel.close()
```
